```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def factory(sigma_square):
    def get_kernel(x, e):
        # sigma_square = np.log(1 / (21 * (1024 / 44100)) + 1)
        sigma = np.sqrt(sigma_square)
        mu = np.log(e) - sigma_square / 2
        result = (
                np.divide(1, x, where=x != 0)
                * sigma * np.sqrt(2 * np.pi)
                * np.exp(
            - ((np.log(x, where=x != 0) - mu) ** 2
               / (2 * sigma ** 2))
        )
        )
        result[0] = 0
        result /= np.sum(result)
        return result

    def convolve(target, kernel, pos):
        # avoid overflow
        left = max(0, pos - 1000)
        right = min(pos + 1000, len(target))
        result = np.zeros(len(target))
        target_w = target[left:right]
        kernel_w = kernel[:right - left]
        result_w = np.convolve(target_w, kernel_w)
        result_w = result_w[:right - left]  # slice to window size
        result[left:right] = result_w
        result /= np.sum(result)
        return result

    def get_expectation(x, y):
        return np.dot(x, y)

    return get_kernel, convolve, get_expectation
```

```python
def test1(axis_resolution, forward_step, nb_iterations, sigma_square):
    # iterate to convolve
    get_kernel, convolve, get_expectation = factory(sigma_square)
    x_axis = np.arange(0, 120, axis_resolution)
    y_axis = np.zeros(len(x_axis))
    y_axis[0] = 1
    position = 0
    pusher = get_kernel(x_axis, forward_step)
    expectation_pusher = get_expectation(x_axis, pusher)
    print('plotting pusher...')
    print('\t theoretical E: %.4f' % forward_step)
    print('\t actual E: %.4f' % expectation_pusher)
    plt.title('pusher')
    plt.plot(x_axis, pusher)
    plt.vlines(expectation_pusher, 0, np.max(pusher), colors='red')
    plt.xlim([0, forward_step * 4])
    plt.show()
    position = 0
    for i in range(nb_iterations):
        # y_axis = convolve(y_axis, pusher, position)
        y_axis = np.convolve(y_axis, pusher)[:len(y_axis)]
        expectation_y_axis = get_expectation(x_axis, y_axis)
        print('plotting #%d iter prob...' % i)
        print('\t theoretical pos: %.4f' % ((i + 1) * forward_step))
        print('\t actual pos: %.4f' % expectation_y_axis)
        plt.title('@ #%d iter prob' % i)
        plt.plot(x_axis, y_axis)
        plt.vlines(expectation_y_axis, 0, np.max(y_axis), colors='red')
        plt.xlim([position - forward_step * 2, position + forward_step * 2])
        plt.show()
        position = round(expectation_y_axis)
```

```python
test1(axis_resolution=256 / 44100,
      forward_step=1024 / 44100,
      nb_iterations=5,
      sigma_square=0.2)
```
